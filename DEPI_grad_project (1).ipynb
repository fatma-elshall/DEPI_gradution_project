```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data cleaning**

```python


# Convert to datetime (since it's in day-first format)
df['Ship Date'] = pd.to_datetime(df['Ship Date'] , dayfirst=True)
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Change the format to MM-DD-YYYY
df['Ship Date'] = df['Ship Date'].dt.strftime('%m-%d-%Y')
df['Order Date'] = df['Order Date'].dt.strftime('%m-%d-%Y')

df.info()
```

```python
# now change ship date , order date to date time format
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.info()
```

```python
df.head()
```

```python
#delete columns('postal code' ,Row id')that won't be important in analysis stage
df.drop(columns=['Postal Code','Row ID'],inplace=True)
df.columns
```

```python
df.shape
```

```python
# chck duplicates
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace =True)
df.shape
```



*   there is no duplicates in data and the data is ready to analysis stage




```python
#check validate dates(ship date before order date )
df[df['Ship Date']< df['Order Date']].value_counts()
```

```python
#remove extra spaces in data
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#save the data after edu=iting
df.to_csv("super store_clean.csv")
```



*   Create visuals for seasonal analysis




```python
#sales trend by quarters

# Set 'Order Date' as index
# important because time series analysis requires the date to be the index for resampling and trend analysis.

df.set_index('Order Date', inplace=True)
```

```python
# Trend and seasonal analysis
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Create a 2x2 grid of plots

# sales trend over years
df.groupby(df['order_year'])['Sales'].sum().plot(kind='line' ,ax=axes[0,0],marker='o',linewidth=2)
axes[0, 0].set_title("Total sales from 2015-2018")
axes[0,0].set_ylabel   #Remove ylabel for clear visualization

#sales trend by quarters
# Set 'Order Date' as index
#important because time series analysis requires the date to be the index for resampling and trend analysis.

#df.set_index('Order Date', inplace=True)

# Resample data quarterly ('Q') and aggregate total sales
df['Sales'].resample('Q').sum().plot(kind='line',marker='o', linestyle='-',ax=axes[0,1])
axes[0,1].set_title("Quarterly sales trend")

#Monthly sales trend comparison by each year

# set  Year & Month as index for resample  trend analysis
df['Year'] = df.index.year
df['Month'] = df.index.month
# Group by Year & Month and sum sales
df_monthly_trend = df.groupby(['Year', 'Month'])['Sales'].sum().unstack(level=0) # to make every year in a separate line

# Plot each year's monthly sales trend
df_monthly_trend.plot(marker='o', linestyle='-',ax=axes[1,0])
axes[1,0].set_title("Monthly Sales Trend Comparison (2015-2018)")
# Formatting the x-axis
axes[1, 0].set_ylabel('Total sales')
axes[1,0].set_xlabel("Month")
plt.xticks([1, 4, 8, 12], ['Jan', 'Apr', 'Aug', 'Dec'])  # Set specific months
plt.grid(True)

#correlatoin between numerical columns
num_df=df[['Month','Year','Sales']]
correlation_matrix=num_df.corr()
#plt.figure(figsize=(8,12))
sns.heatmap(correlation_matrix , annot=True , cmap="coolwarm", fmt=".2f", linewidths=0.5,ax=axes[1,1])
axes[1,1].set_title("correlation map between numerical values")
axes[1,1].set_xlabel("")
axes[1,1].set_ylabel("")
#plt.title("correlation map between numerical values")
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

**Some comments after analysis**



*   It is noticeable that the number of orders increases gradually each year, but this does not necessarily mean that total sales are higher, as it depends on the quantity and price of each product ordered.

* Sales increase in "March" ,"September" and become highst peak at "November" ِalso sales deacrsed in "February" and "October" in each year . I believe this increase is due to the beginning of a new season, such as summer, sales increase as customer purchase more items .
* Additionally, in September, the start of a new school year contributes to higher sales, while in November, the rise is driven by White Friday discounts.

* There is no relation between date and sales!

**Create dashboard with the most important information**

```python
# summary dashboard

# Define the color palette
custom_color = "#D84040"
pie_colors = ["#D84040", "#A31D1D", "#ECDCBF", "#F8F2DE"]  # Pie chart colors

fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # Create a 2×3 grid of plots

## Trend Seasonal Analysis
df_monthly_trend.plot(marker='o', linestyle='-', ax=axes[0, 0])
axes[0, 0].set_title("Sales Trend Comparison (2015-2018)")
axes[0, 0].set_ylabel("Total Sales")
axes[0, 0].set_xlabel("Month")

## Customer Analysis
df.groupby("Customer Name")["Sales"].sum().nlargest(10).plot(kind="bar", color=custom_color, ax=axes[0, 1])
axes[0, 1].set_title("Top 10 Customers by Sales", color=custom_color)

## Product Analysis
df.groupby("Category")["Sales"].sum().sort_values(ascending=False).plot(kind="bar", color=custom_color, ax=axes[0, 2])
axes[0, 2].set_title("Revenue by Category", color=custom_color)
axes[0, 2].set_ylabel("Total Sales", color=custom_color)

df.groupby("Sub-Category")["Sales"].sum().nlargest(10).plot(kind="bar", color=custom_color, ax=axes[1, 0])
axes[1, 0].set_title("Top 10 Subcategories in Sales", color=custom_color)

## Revenue by customer segment
df.groupby("Segment")["Sales"].sum().plot(kind="pie", autopct="%.2f%%", colors=pie_colors, ax=axes[1, 1])  # Fixed "color" to "colors"
axes[1, 1].set_title("Revenue by Customer Segment", color=custom_color)
axes[1, 1].set_ylabel("")  # Remove y-label for cleaner visualization

## Sales by state
df.groupby("State")["Sales"].sum().nlargest(4).plot(kind="pie", autopct="%.2f%%", colors=pie_colors, ax=axes[1, 2])  # Fixed "color" to "colors"
axes[1, 2].set_title("Sales by State", color=custom_color)
axes[1, 2].set_ylabel("")  # Remove y-label for cleaner visualization

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
